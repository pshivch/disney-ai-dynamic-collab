# Disney AI: Dynamic Collaboration 🎬✨
This notebook generates advanced visuals + game theory math for the Disney (Fantasia-inspired) mirror.

```

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
from itertools import product

# Make plots look nice
sns.set(style="whitegrid", palette="muted")
```

## Character Frequency (Bar Chart)

```

characters = ["Mickey", "Elsa", "Woody", "Moana", "Genie", "Baymax", "Ariel"]
scene_counts = [35, 55, 47, 49, 37, 15, 29]

plt.figure(figsize=(8,5))
sns.barplot(x=characters, y=scene_counts, color="orange")
plt.title("Character Presence in Collaborative Scenes")
plt.ylabel("Scene Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("character_frequency.png")
plt.show()
```

## Collaboration Heatmap (Agent Influence)

```

np.random.seed(42)
agents = ["Mickey","Elsa","Woody","Moana","Genie"]
data = np.random.rand(len(agents), len(agents))

plt.figure(figsize=(6,5))
sns.heatmap(data, xticklabels=agents, yticklabels=agents, cmap="viridis", annot=True)
plt.title("Collaboration Heatmap (Agent Influence)")
plt.tight_layout()
plt.savefig("collaboration_heatmap.png")
plt.show()
```

## Tempo Shifts: AI vs Human (Line Chart)

```

time = np.linspace(0, 180, 200)
human = 100 + 10*np.sin(time/15)
ai = 100 + 12*np.sin(time/15 + 1)

plt.figure(figsize=(10,5))
plt.plot(time, human, label="Human", color="orange")
plt.plot(time, ai, label="AI", color="blue")
plt.title("Tempo Shifts: AI vs Human Suggestion")
plt.xlabel("Time (s)")
plt.ylabel("Tempo (BPM)")
plt.legend()
plt.tight_layout()
plt.savefig("tempo_ai_vs_human.png")
plt.show()
```

## Shapley Values (Game Theory)

```

players = ["Mickey","Elsa","Woody"]
values = {"Mickey":0.4,"Elsa":0.35,"Woody":0.25}

plt.figure(figsize=(6,5))
sns.barplot(x=list(values.keys()), y=list(values.values()), palette="muted")
plt.title("Shapley Value Contributions")
plt.ylabel("Value Share")
plt.ylim(0,0.5)
plt.tight_layout()
plt.savefig("shapley_values.png")
plt.show()
```

## Workflow Diagram (Matplotlib)

```

fig, ax = plt.subplots(figsize=(8,5))

steps = ["Story Prompt", "Agent Negotiation (Game Theory)", "Scene Plan",
         "Music Alignment", "Animation Blocking", "Dialogue Timing", "Rendered Preview"]

# Coordinates
coords = {
    "Story Prompt": (0.1, 0.8),
    "Agent Negotiation (Game Theory)": (0.4, 0.8),
    "Scene Plan": (0.7, 0.8),
    "Music Alignment": (0.3, 0.5),
    "Animation Blocking": (0.5, 0.5),
    "Dialogue Timing": (0.7, 0.5),
    "Rendered Preview": (0.5, 0.2)
}

for step, (x,y) in coords.items():
    ax.text(x, y, step, ha="center", va="center",
            bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue", alpha=0.5))

# Arrows
ax.annotate("", xy=coords["Agent Negotiation (Game Theory)"], xytext=coords["Story Prompt"],
            arrowprops=dict(arrowstyle="->"))
ax.annotate("", xy=coords["Scene Plan"], xytext=coords["Agent Negotiation (Game Theory)"],
            arrowprops=dict(arrowstyle="->"))
ax.annotate("", xy=coords["Music Alignment"], xytext=coords["Scene Plan"],
            arrowprops=dict(arrowstyle="->"))
ax.annotate("", xy=coords["Animation Blocking"], xytext=coords["Scene Plan"],
            arrowprops=dict(arrowstyle="->"))
ax.annotate("", xy=coords["Dialogue Timing"], xytext=coords["Scene Plan"],
            arrowprops=dict(arrowstyle="->"))
for prev in ["Music Alignment","Animation Blocking","Dialogue Timing"]:
    ax.annotate("", xy=coords["Rendered Preview"], xytext=coords[prev],
                arrowprops=dict(arrowstyle="->"))

ax.axis("off")
plt.title("Orchestration Workflow (Disney AI: Dynamic Collaboration)")
plt.tight_layout()
plt.savefig("workflow_diagram.png")
plt.show()
```
