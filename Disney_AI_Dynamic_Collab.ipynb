# Disney AI: Dynamic Collaboration (Fantasia-Inspired)
This notebook demonstrates advanced components: Shapley values, Nash bargaining, and visuals.

```
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng(42)
```

```

characters = ["Mickey","Elsa","Woody","Moana","Genie","Baymax","Ariel"]
presence = np.array([34, 56, 48, 40, 36, 15, 22])

plt.figure()
plt.bar(range(len(characters)), presence)
plt.xticks(range(len(characters)), characters)
plt.title("Character Presence in Collaborative Scenes")
plt.xlabel("Character"); plt.ylabel("Scene Count")
plt.tight_layout(); plt.show()
```

```

import itertools, math

sections = ["Strings","Winds","Brass","Percussion"]
base = {s: rng.uniform(0.5,1.0) for s in sections}

def v(coalition):
    if not coalition: return 0.0
    score = sum(base[s] for s in coalition)
    if "Strings" in coalition and "Winds" in coalition: score += 0.3
    if "Brass" in coalition and "Percussion" in coalition: score += 0.2
    return score

def shapley(players):
    n=len(players); sv={p:0.0 for p in players}
    for perm in itertools.permutations(players):
        S=set()
        for p in perm:
            before=v(tuple(S)); S.add(p); after=v(tuple(S))
            sv[p]+=after-before
    return {k: sv[k]/math.factorial(n) for k in players}

sv = shapley(sections)
plt.figure(); plt.bar(range(len(sections)), [sv[s] for s in sections])
plt.xticks(range(len(sections)), sections)
plt.title("Shapley Values: Section Contribution")
plt.tight_layout(); plt.show()
sv
```

```

T = 180; time = np.arange(T)
human_tempo = 100 + 10*np.sin(time/20.0) + rng.normal(0,0.8,T)
ai_tempo    = 105 + 8*np.sin((time+10)/22.0) + rng.normal(0,0.8,T)

plt.figure(); plt.plot(time,human_tempo,label="Human")
plt.plot(time,ai_tempo,label="AI")
plt.title("Tempo Shifts: AI vs Human"); plt.xlabel("Time (s)"); plt.ylabel("Tempo (BPM)")
plt.legend(); plt.tight_layout(); plt.show()
```

```

agents = ["Mickey","Elsa","Woody","Moana","Genie"]
M = rng.uniform(0,1,(len(agents), len(agents)))
for i in range(len(agents)): M[i,i]=1.0

plt.figure(); im=plt.imshow(M,aspect='auto')
plt.xticks(range(len(agents)), agents); plt.yticks(range(len(agents)), agents)
plt.title("Collaboration Heatmap (Agent Influence)")
plt.colorbar(im, fraction=0.046, pad=0.04)
plt.tight_layout(); plt.show()
```
